# **Criar o Dataframe dos placares**

```python
import pandas as pd
#utilizando a biblioteca pandas
df_jogos = pd.read_csv("dados_brasileirao/campeonato-brasileiro-full.csv")
#visualizar os nomes das colunas e numero
df_jogos.info()
```

## Dados faltando

```python
df_jogos.isnull().sum()
```

```python
df_jogos = df_jogos.drop(columns=['arrecadacao'])
```

Como a arrecadaçâo está faltando na maioria dos jogos, vou remove-la e fazer um estudo se os dados de formação e técnico faltantes são mais novos ou mais antigos, já que eles seriam indispensaveis para uma analise mais aprofundada.

```python
df_jogos['data'] = pd.to_datetime(df_jogos['data'])
#transformei os textos de data em formato datetime
df_jogos['ano'] = df_jogos['data'].dt.year
#criei uma nova coluna chamada ano

df_jogos.groupby('ano')[['tecnico_mandante', 'tecnico_visitante', 'formacao_visitante','formacao_mandante']].apply(lambda x: x.isnull().mean()*100 )
#calculei a % de valores ausentes por ano
```

A coluna de data foi convertida para o formato datetime, o que permitiu criar uma coluna com o ano. Com isso, foi possível analisar os valores ausentes ao longo dos anos. Observei que os dados de técnicos estão disponíveis a partir de 2014, enquanto as informações de formação das equipes só passam a ter uma boa quantidade de dados a partir de 2015. Por isso, optei por utilizar os dados a partir de 2015, garantindo maior consistência na análise.
